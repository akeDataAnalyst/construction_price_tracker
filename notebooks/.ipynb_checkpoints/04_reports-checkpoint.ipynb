#### Report Generation

#### Goals:
- Summary statistics & visualizations in notebook
- Automated PDF report (tables + charts)
- Focus on construction/interior pricing insights

#### 1. Load the cleaned data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
from datetime import datetime

# better looking plots
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("muted")

# Load the cleaned data 

DATA_DIR = Path("data")
CLEAN_FILE = DATA_DIR / "master_prices_cleaned_20260218.csv"  

df = pd.read_csv(CLEAN_FILE, parse_dates=['last_checked_dt', 'scraped_at'])

print(f"Loaded {len(df)} cleaned rows")
print("Columns:", df.columns.tolist())
df.head(3)
```

#### 2. Basic Summary Statistics

##### Items & prices per category

```python
summary_cat = df[df['price_etb_valid']].groupby('category').agg(
    items=('material', 'count'),
    avg_price=('price_etb', 'mean'),
    median_price=('price_etb', 'median'),
    min_price=('price_etb', 'min'),
    max_price=('price_etb', 'max'),
    outdated_pct=('is_outdated', 'mean')
).round(2)

summary_cat['outdated_pct'] *= 100
summary_cat = summary_cat.sort_values('avg_price', ascending=False)

print("Summary by Category")
display(summary_cat)
```

##### Top 10 most expensive materials (valid prices only)

```python
top_expensive = df[df['price_etb_valid']].nlargest(10, 'price_etb')[
    ['category', 'material_clean', 'price_etb', 'unit_standard', 'last_checked_dt']
]
print("\nTop 10 most expensive materials:")
display(top_expensive)
```

##### Freshness overview

```python
freshness = df['is_outdated'].value_counts(normalize=True) * 100
print("\nData freshness:")
print(f"Fresh (<= 365 days): {freshness.get(False, 0):.1f}%")
print(f"Outdated (>365 days): {freshness.get(True, 0):.1f}%")
```

#### 3. Visualizations

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10), sharex=False)

# A. Average price per category (bar)
sns.barplot(
    data=summary_cat.reset_index(),
    x='avg_price',
    y='category',
    ax=axes[0,0]
)
axes[0,0].set_title('Average Price per Category (ETB)')
axes[0,0].set_xlabel('Average Price (ETB)')

# B. Price distribution (boxplot - log scale for readability)
sns.boxplot(
    data=df[df['price_etb_valid']],
    x='price_etb',
    y='category',
    ax=axes[0,1],
    showfliers=False
)
axes[0,1].set_xscale('log')
axes[0,1].set_title('Price Distribution by Category (log scale)')
axes[0,1].set_xlabel('Price (ETB)  log scale')

# C. Outdated percentage
summary_cat['outdated_pct'].plot(kind='barh', ax=axes[1,0])
axes[1,0].set_title('% Outdated Items per Category')
axes[1,0].set_xlabel('Outdated %')

# D. Scatter: price vs days since update
sns.scatterplot(
    data=df[df['price_etb_valid']],
    x='days_since_update',
    y='price_etb',
    hue='category',
    alpha=0.6,
    ax=axes[1,1]
)
axes[1,1].set_yscale('log')
axes[1,1].set_title('Price vs Age of Data Point')
axes[1,1].set_xlabel('Days since last checked')
axes[1,1].set_ylabel('Price (ETB)  log scale')

plt.tight_layout()
plt.show()
```

#### 4. PDF Report Generation

```python
from reportlab.lib.pagesizes import A4
from reportlab.platypus import (
    SimpleDocTemplate, Paragraph, Spacer, Table, TableStyle,
    Image as RLImage
)
from reportlab.lib.styles import getSampleStyleSheet, ParagraphStyle
from reportlab.lib import colors
from reportlab.lib.units import cm
from io import BytesIO

# save matplotlib figure to bytes
def fig_to_bytes(fig):
    buf = BytesIO()
    fig.savefig(buf, format='png', dpi=150, bbox_inches='tight')
    buf.seek(0)
    return buf

# Create PDF
pdf_path = DATA_DIR / f"construction_prices_report_{datetime.now().strftime('%Y%m%d')}.pdf"
doc = SimpleDocTemplate(str(pdf_path), pagesize=A4, rightMargin=2*cm, leftMargin=2*cm, topMargin=2*cm, bottomMargin=2*cm)

styles = getSampleStyleSheet()
title_style = ParagraphStyle('Title', parent=styles['Title'], fontSize=18, spaceAfter=12)
heading_style = ParagraphStyle('Heading2', parent=styles['Heading2'], spaceAfter=8)
normal_style = styles['Normal']

story = []

# Header
story.append(Paragraph("Construction Materials Price Tracker Report", title_style))
story.append(Paragraph(f"Generated: {datetime.now().strftime('%Y-%m-%d %H:%M EAT')}", normal_style))
story.append(Spacer(1, 0.8*cm))

# 1. Summary table
story.append(Paragraph("1. Summary by Category", heading_style))

table_data = [summary_cat.reset_index().columns.tolist()] + summary_cat.reset_index().values.tolist()
t = Table(table_data)
t.setStyle(TableStyle([
    ('BACKGROUND', (0,0), (-1,0), colors.grey),
    ('TEXTCOLOR', (0,0), (-1,0), colors.whitesmoke),
    ('ALIGN', (0,0), (-1,-1), 'CENTER'),
    ('FONTNAME', (0,0), (-1,0), 'Helvetica-Bold'),
    ('BOTTOMPADDING', (0,0), (-1,0), 12),
    ('BACKGROUND', (0,1), (-1,-1), colors.beige),
    ('GRID', (0,0), (-1,-1), 1, colors.black),
    ('VALIGN', (0,0), (-1,-1), 'MIDDLE'),
]))
story.append(t)
story.append(Spacer(1, 1*cm))

# 2. Charts
story.append(Paragraph("2. Key Visualizations", heading_style))

# Chart 1  Avg price bar
fig1 = plt.figure(figsize=(8, 4.5))
sns.barplot(data=summary_cat.reset_index(), x='avg_price', y='category')
plt.title('Average Price per Category (ETB)')
plt.xlabel('Average Price (ETB)')
plt.tight_layout()
story.append(RLImage(fig_to_bytes(fig1), width=16*cm, height=9*cm))
plt.close(fig1)
story.append(Spacer(1, 0.6*cm))

# Chart 2  Price distribution boxplot
fig2 = plt.figure(figsize=(8, 4.5))
sns.boxplot(data=df[df['price_etb_valid']], x='price_etb', y='category', showfliers=False)
plt.xscale('log')
plt.title('Price Distribution by Category (log scale)')
plt.xlabel('Price (ETB)  log scale')
plt.tight_layout()
story.append(RLImage(fig_to_bytes(fig2), width=16*cm, height=9*cm))
plt.close(fig2)

# Footer note
story.append(Spacer(1, 1.2*cm))
story.append(Paragraph("Data source: 2merkato.com (scraped for educational/portfolio purposes)", normal_style))
story.append(Paragraph(f"Total valid items: {df['price_etb_valid'].sum()}", normal_style))

# Add footer
def add_footer(canvas, doc):
    canvas.saveState()
    canvas.setFont("Helvetica", 9)
    canvas.setFillColor(colors.grey)
    canvas.drawString(2*cm, 1*cm, "Developed by Aklilu Abera  Construction Materials Price Tracker")
    canvas.restoreState()

# Build PDF
doc.build(story, onFirstPage=add_footer, onLaterPages=add_footer)
print(f"PDF report generated  {pdf_path}")
```
